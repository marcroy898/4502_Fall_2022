```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import yfinance as yf
```

```python
# Choose Company
BP = yf.Ticker("BP")
XEL = yf.Ticker("XEL")
SHEL = yf.Ticker("SHEL")
```

```python
# get historical market data
start = "2016-01-01"
end = "2016-12-31"
interval = '1d'
dataBP = BP.history(period="max",
    interval = interval,
    start = start,
    end = end
)
dataXEL = XEL.history(period="max",
    interval = interval,
    start = start,
    end = end
)
dataSHEL = SHEL.history(period="max",
    interval = interval,
    start = start,
    end = end
)
marker = '2016-05-12'
```

```python
# 12 May 2016, Shell Oil Spill
```

```python
dataBP['Volume'].plot()
plt.axvline(pd.Timestamp(marker),color='r')
```

```python
dataXEL['Volume'].plot()
plt.axvline(pd.Timestamp(marker),color='r')
```
